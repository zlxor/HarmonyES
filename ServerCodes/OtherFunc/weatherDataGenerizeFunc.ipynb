```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data.csv")
```

```python
df.describe()
```

```python
length = df.shape[0]
```

```python
pv = []
for hour in range(0, length) :
    x = 0
    x = x + (df['temperature'][hour] - 10) / 21 * 6
    x = x + (df['light_intensity'][hour]) / 3000000 * 10
    x = x + 6 + (df['wind_speed'][hour]) / 5 * (-3)
    x = x + 3 + (df['humidity'][hour] - 23) / 77 * (-3)
    if x < 0 :
        x = 0
    pv.append(x)
```

```python
df2 = pd.DataFrame(pv, columns = ["solar_power"])
```

```python
df2.describe()
```

```python
wd = []

def cub(x):
    return x * x * x

for hour in range(0, length) :
    x = 0
    x = x + 7 - (df['temperature'][hour] - 10) / 21 * 6
    x = x + 3 - (df['light_intensity'][hour]) / 3000000 * 3
    x = x + cub(df['wind_speed'][hour]) / 55 * 10
    x = x + (df['humidity'][hour] - 23) / 77 * 6
    if x < 0 :
        x = 0
    wd.append(x)
```

```python
df3 = pd.DataFrame(wd, columns = ["wind_power"])
```

```python
df3.describe()
```

```python
cs = []
for hour in range(0, length) :
    x = 0
    x = x + (df['temperature'][hour] - 10) / 21 * 25
    x = x + (df['light_intensity'][hour]) / 3000000 * 20
    x = x + 5 + (df['wind_speed'][hour]) / 5 * (-5)
    x = x + 5 + (df['humidity'][hour] - 23) / 77 * (-5)
    if x < 0 :
        x = 0
    cs.append(-x)
```

```python
df4 = pd.DataFrame(cs, columns = ["consumption"])
```

```python
df4.describe()
```

```python
df["solar_power"] = pv
df["wind_power"] = wd
df["consumption"] = cs
```

```python
df.head(10)
```

```python
df.to_csv("weather_data_train.csv", index = False)
```
